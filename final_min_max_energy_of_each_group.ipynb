```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dfs = pd.read_excel('US_regions.xlsx', sheet_name="USA Locations")
dfs
```

```python
dfs = dfs.drop(dfs.columns[[-1,-2,-3,-4,-5,-6,-7,-8, 2, 3]], axis=1)
dfs
```

```python
dfs = dfs[dfs['Unnamed: 1'].notnull()]
dfs
```

```python
dfs.columns = ['LOCATION', 'ID', 'CZ']
dfs
```

```python
dfs = dfs.iloc[1:]
dfs
```

```python
dfs = dfs.reset_index(drop=True)
dfs
```

```python
df = pd.read_csv('heating_cooling_with_id.csv', delimiter = ',')
df
```

```python
df_merge = pd.merge(dfs, df, on=['ID'], how='right')
df_merge
```

```python
df_merge = df_merge[df_merge['LOCATION'].notnull()]
df_merge
```

```python
df_merge['Total Energy'] = df_merge['Heating'] + df_merge['Cooling']
df_merge
```

```python
df_merge = df_merge.sort_values(['CZ'])
df_merge
```

```python
df_merge_max = df_merge.groupby(['CZ'], as_index=False)['Total Energy'].max()
df_merge_max.columns = ['CZ1', 'Max energy']
df_merge_max
```

```python
df_merge_min = df_merge.groupby(['CZ'], as_index=False)['Total Energy'].min()
df_merge_min.columns = ['CZ', 'Min energy']
df_merge_min
```

```python
df1 = pd.concat([df_merge_min, df_merge_max], axis=1, sort=False)
df1 = df1.drop(df1.columns[[-2]], axis=1)
df1.columns = ['CZ', 'Min_energy', 'Max_energy']
df1['Min_energy'] = df1['Min_energy']/100
df1['Max_energy'] = df1['Max_energy']/100
df1.Min_energy = df1.Min_energy.astype(int)
df1.Max_energy = df1.Max_energy.astype(int)
#df1.set_index("CZ",drop=True,inplace=True)
dfToList = df1['CZ'].tolist()
df1
```

```python
#df1.plot(x="CZ", y=["Min energy", "Max energy"])
y_pos = np.arange(len(dfToList))
df1.plot(x="CZ", y=["Min_energy", "Max_energy"], style='.-')
plt.ylim(-100, 3000)
plt.xticks(y_pos, dfToList)
#ax.set_xticklabels(df1.CZ)
```

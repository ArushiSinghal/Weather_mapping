```python
import pandas as pd
```

```python
dfs = pd.read_excel('US_regions.xlsx', sheet_name="USA Locations")
dfs
```

```python
dfs = dfs.drop(dfs.columns[[-1,-2,-3,-4,-5,-6,-7,-8, 2, 3]], axis=1)
dfs
```

```python
dfs = dfs[dfs['Unnamed: 1'].notnull()]
dfs
```

```python
dfs.columns = ['LOCATION', 'ID', 'CZ']
dfs
```

```python
dfs = dfs.iloc[1:]
dfs
```

```python
dfs = dfs.reset_index(drop=True)
dfs
```

```python
df = pd.read_csv('heating_cooling_with_id.csv', delimiter = ',')
df
```

```python
df_merge = pd.merge(dfs, df, on=['ID'], how='right')
df_merge
```

```python
df_merge = df_merge[df_merge['LOCATION'].notnull()]
df_merge
```

```python
pd.options.mode.chained_assignment = None
df_merge['Total Energy'] = df_merge['Heating'] + df_merge['Cooling']
print (type(df_merge))
df_merge.to_csv('a7.csv', sep='\t')
df_merge
#df_merge.to_csv('a8.csv', sep='\t', encoding='utf-8')
```

```python
df_merge = df_merge.sort_values(['CZ'])
df_merge
```
